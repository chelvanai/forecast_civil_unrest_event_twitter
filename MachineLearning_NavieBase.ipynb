```python
def clear_punctuation(s):
    clear_string = ""
    for symbol in s:
        if symbol not in string.punctuation:
            clear_string += symbol
    return clear_string
```

```python
string_punctuation = ".#http,_;xf0x9fbRT"
def remove_punctuation(s):
    no_punct = ""
    for letter in s:
        if letter not in string_punctuation:
            no_punct += letter
    return no_punct
```

```python
import nltk.classify.util
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import names
```

```python
def word_feats(words):
    return dict([(word, True) for word in words])
```

```python
civil_unreset = ['protest', 'strike', 'objection', 'strife', 'conflict', 'fight', 'quarrel','wrangle ']
religion_conflict = ['islam', 'Buddhist', 'hindu', 'hate']
neutral_tweet = ['area', 'book', 'business', 'case', 'child', 'company', 'country', 'day', 'eye', 'fact', 'family', 'government', 'group', 'hand', 'home', 'job', 'life', 'lot', 'man', 'money', 'month', 'mother', 'Mr', 'night', 'number', 'part', 'people', 'place', 'point', 'problem', 'program', 'question', 'right', 'room', 'school', 'state', 'story', 'student', 'study', 'system', 'thing', 'time', 'water', 'way', 'week', 'woman', 'word', 'work', 'world', 'year']
```

```python
civil_unreset_features = [(word_feats(civil), 'civil') for civil in civil_unreset]
religion_conflict_features = [(word_feats(religion), 'religion') for religion in religion_conflict]
neutral_tweet_features = [(word_feats(neu), 'neutral') for neu in neutral_tweet]
```

```python
train_set = civil_unreset_features + neutral_tweet_features + religion_conflict_features
```

```python
classifier = NaiveBayesClassifier.train(train_set)
```

```python
def find(sentence):
    sentence = sentence.lower()
    civil = 0
    religion = 0
    netural = 0
    words = sentence.split(' ')

    for word in words:
    
        classResult = classifier.classify(word_feats(word))
        if classResult == 'civil':
            civil = civil+1
        if classResult == 'religion':
            religion = religion+1
        if classResult == 'neutral':
            netural = netural+1
        
    print("civil unrest:- "+str(float(civil) / len(words)))
    print("religion conflict:- "+str(float(religion) / len(words)))
    print("neutral tweet:- "+str(float(netural) / len(words)))
```

```python
for i in data:
    s = remove_punctuation(clear_punctuation(i))
    print(s)
    find(s)
    print("")
```
